```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load your DataFrame (replace with your actual data source)
# df = pd.read_csv('your_data.csv')

# For demonstration, assume df is already defined in your environment
nutri_cols = [
    'Eat at an on-campus eatery (Food trucks, restaurants, food court)',
    'Prepared a meal on my own',
    'Eat with friends',
    'Eat alone'
]

# 2. One-hot encode the nutritional columns
encoded = pd.get_dummies(df[nutri_cols])
encoded_df = pd.concat([df, encoded], axis=1)

# 3. Plot individual heatmaps for the first 20 rows of each encoded variable
for col in nutri_cols:
    encoded_cols = [c for c in encoded_df.columns if c.startswith(col + '_')]
    data = encoded_df[encoded_cols].head(20)
    
    plt.figure(figsize=(6, 4))
    plt.imshow(data, aspect='auto', interpolation='nearest')
    plt.colorbar(label='0 or 1')
    
    plt.xticks(range(len(encoded_cols)), encoded_cols, rotation=45, ha='right')
    plt.yticks(range(20), range(1, 21))
    
    plt.title(f"Heatmap of '{col}' Encoding (First 20 Rows)")
    plt.xlabel("Category")
    plt.ylabel("Sample Index")
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Create bar charts for each nutritional column
nutri_cols = [
    "Eat at an on campus eatery (Food trucks, restaurants, food court)",
    "Prepared a meal on my own", 
    "Eat with friends",
    "Eat alone"
]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for idx, col in enumerate(nutri_cols):
    value_counts = df[col].value_counts()
    axes[idx].bar(value_counts.index, value_counts.values)
    axes[idx].set_title(col)
    axes[idx].set_xlabel('Response')
    axes[idx].set_ylabel('Count')
    axes[idx].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

nutri_cols = [
    "Eat at an on campus eatery (Food trucks, restaurants, food court)",
    "Prepared a meal on my own",
    "Eat with friends",
    "Eat alone"
]

encoded_nutri = pd.get_dummies(df[nutri_cols], prefix_sep="_")

print("New one-hot columns:")
print(encoded_nutri.columns.tolist())

plt.figure(figsize=(14,6))
sns.heatmap(
    encoded_nutri.head(20),
    annot=True,
    cbar=True,
    cmap="YlGnBu",
    linewidths=0.5
)
plt.xticks(rotation=90)
plt.title("Nutritional Questions One-Hot (First 20 Samples)")
plt.xlabel("Encoded Category")
plt.ylabel("Sample Index")
plt.tight_layout()
plt.show()
```

```python
for q in nutri_cols:
    cols = [c for c in encoded_nutri.columns if c.startswith(q + "_")]
    sns.heatmap(encoded_nutri[cols].head(20), annot=True, cbar=False,
                cmap="YlGnBu", linewidths=.5)
    plt.title(q)
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    plt.show()
```
